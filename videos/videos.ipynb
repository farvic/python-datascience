```python
import pandas as pd

# DF was sorted by views in excel
videos_df = pd.read_csv('videos.csv', encoding='unicode_escape', index_col=0)

videos_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# get the top 5 videos
top_5 = videos_df.iloc[0:5, -1]

# process the titles
titles = list(top_5.index[0:5])
titles = [x[0:30] for x in titles]
titles[1] = 'Aquaponia passo a passo'
titles[2] = 'Fossa séptica biodigestora'


# plot the top 5 videos as a barplot
plt.title('Top 5 videos by views')
sns.barplot(y=titles, x=top_5.values, width=0.5)
plt.xticks(rotation=45)
plt.show()
```
