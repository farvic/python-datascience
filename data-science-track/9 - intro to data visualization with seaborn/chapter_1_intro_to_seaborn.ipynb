```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

countries = pd.read_csv('countries-of-the-world.csv')

gdp = countries['GDP ($ per capita)'].to_list()

phones = countries['Phones (per 1000)'].to_list()


# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=phones)

# Show plot
plt.show()
```

### Countplot on the y-axis

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

countries = pd.read_csv('countries-of-the-world.csv')

region = countries['Region'].to_list()

# Create count plot with region on the y-axis
sns.countplot(y=region)
plt.title('Countries of the World by Region')

# Show plot
plt.show()
```

### Tips dataset
Example dataset from seaborn's package

```python
import pandas as pd
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()

sns.scatterplot(x="total_bill",y="tip",data=tips, hue="smoker")
plt.show()
```

### Customizing hue colors according to the labels

```python
import pandas as pd
import seaborn as sns
hue_colors = {"Yes":"black",
							"No":"red"}
tips = sns.load_dataset("tips")


sns.scatterplot(x="total_bill",y="tip",data=tips, hue="smoker",palette=hue_colors)
plt.show()
```

### hue with countplots

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

sns.countplot(x="smoker",data=tips,hue="sex")

plt.show()
```

### Putting all together

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

student_data = pd.read_csv('student-alcohol-consumption.csv')

# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(data=student_data, x="school",hue="location",palette=palette_colors)

# Display plot
plt.show()
```
